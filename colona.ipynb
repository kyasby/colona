```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pytablewriter

writer = pytablewriter.MarkdownTableWriter()
```

```python
df=pd.read_csv("COVID-19.csv")
df
```

```python
df = df[["受診都道府県", "居住都道府県", "人数", "発症日", "確定日"]].rename(columns={"年代":"age", "性別":"sex", "受診都道府県":"hsp", "居住都道府県":"house"})
df
```

```python
for i, row in df.iterrows():
    if type(row["発症日"])==float:
        df.at[i, "発症日"] = row["確定日"]
    else:
        pass
```

```python
todofuken = pd.read_csv("japan.csv", header=None)[0]
todofuken
```

```python
df["hsp"].value_counts()
```

```python
df["hsp"]= df["hsp"].apply(lambda x : "その他" if x not in list(todofuken) else x)
        
```

```python
df
```

```python
df = df.rename(columns = {"発症日":"Date"})
```

```python
jpn = pd.read_csv("japan.csv", header=None)
```

```python
pvt = df.pivot_table(index="hsp", columns="Date", values="人数").rename(index = dict(zip(jpn[0], jpn[2]))).fillna(0).rename(index={"その他":"others"})
```

```python
pvt
```

```python
for i in range(len(pvt)):
    pvt.iloc[i]=pvt.iloc[i].cumsum()
```

```python
pvt.iloc[:,20:]
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(pvt.iloc[:,-60:] , linewidths=0, cmap='Spectral', cbar=True, xticklabels=5)
plt.savefig("colona.png")
```
